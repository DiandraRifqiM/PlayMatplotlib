```python
import matplotlib.pyplot as plt
import numpy as np

xpts = np.array([0, 10])
ypts = np.array([0, 125])

plt.plot( xpts, ypts)
plt.show()
```

```python
# Plot Style
plt.style.use('bmh')
plt.tight_layout()

ages_x = np.arange(25, 35) 
NG_Engineer = ([38500, 41200, 43950, 47300, 50550, 53800, 59200, 63600, 68950, 75000])
LM_Engineer = ([45800, 48650, 51200, 54750, 57900, 61300, 64950, 69500, 73400, 85600])

# Create title
plt.title("Average Engineer Salary (USD) per 5 Month by Ages")
plt.xlabel("Ages")
plt.ylabel("Salary")

# Create plot & legend
plt.plot( ages_x, NG_Engineer, color='#0525b3', label="Northrop Grumman Engineer")
plt.plot( ages_x, LM_Engineer, color='#b30505', label="Lockhead Martin Engineer")
plt.legend()

# Save Plot
plt.savefig('PLot.png')

plt.show()

# print(plt.style.available)
```
