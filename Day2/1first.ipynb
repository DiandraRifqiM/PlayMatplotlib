```python
import matplotlib.pyplot as plt
import pandas as pd
# from collections import Counter
import numpy as np

plt.style.use('ggplot')

ages_x = np.arange(25, 35) 
NG_Engineer = ([38500, 41200, 43950, 47300, 50550, 53800, 59200, 63600, 68950, 75000])
LM_Engineer = ([45800, 48650, 51200, 54750, 57900, 61300, 64950, 69500, 73400, 85600])

```

```python
# Bar Charts
x_indexes = np.arange(len(ages_x))
width = 0.3
plt.bar(x_indexes - width/2, NG_Engineer, width=width, color='red', label='Northrop Grumman Engineer')
plt.bar(x_indexes + width/2, LM_Engineer, width=width, color='blue', label='Lockhead Martin Engineer')

# Title
plt.legend()
plt.title("Average Engineer Salary (USD) per 5 Month by Ages")
plt.xlabel("Ages")
plt.ylabel("Salary")
plt.xticks(ticks=x_indexes, labels=ages_x)
```

```python
# CSV File
df = pd.read_csv(r"D:\Python Course\Playing Matplotlib\csv\data.csv")
df["LanguagesWorkedWith"] = df["LanguagesWorkedWith"].str.split(';')
df = df.explode('LanguagesWorkedWith')
df = df.reset_index(drop= True)
df[:5]
    
```

```python

NumUsers = df["LanguagesWorkedWith"].value_counts().sort_values(ascending=True)

plt.figure(figsize=(10, 8))
plt.barh(NumUsers.index, NumUsers.values, color="#0ae5f5", height=0.9)
plt.xlabel("Number of People Uses")
plt.title("Top 20 Programming Languages")
plt.tight_layout()
plt.legend()
plt.show()
```
