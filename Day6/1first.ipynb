```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('bmh')
```

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
          538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.figure(figsize=(12, 9))
plt.scatter(x, y, c=colors, s=sizes, cmap='cool', alpha=0.9, edgecolors='black', linewidths=0.5)
cbar = plt.colorbar()
cbar.set_label("Satisfcation")
```

```python
df = pd.read_csv(r"D:\Python Course\Playing Matplotlib\csv\2019-05-31-data.csv")
vc = df["view_count"]
lik = df["likes"]
rat = df["ratio"]

# plt.figure(figsize=(12, 9))

plt.scatter(vc, lik, c=rat, cmap='cool', linewidths=0.5, alpha=0.9, edgecolors='black')

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label("Like & Dislike Ratio")

plt.title("Trending Youtube Video")
plt.xlabel("View Count")
plt.ylabel("Total Likes")

plt.tight_layout()
plt.show()
```
