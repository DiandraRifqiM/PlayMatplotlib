```python
import pandas as pd
import matplotlib.pyplot as plt

# Chart Style
plt.style.use('fivethirtyeight')

df = pd.read_csv(r"D:\Python Course\Playing Matplotlib\csv\data.csv")
df["LanguagesWorkedWith"] = df["LanguagesWorkedWith"].str.split(';')
df = df.explode("LanguagesWorkedWith")
df.reset_index(drop=True)
df.head(5)
```

```python
NumUser = df["LanguagesWorkedWith"].value_counts().head(7)

plt.figure(figsize=(12, 10))
plt.pie(NumUser.values, labels=NumUser.index, wedgeprops={'edgecolor':'black'}, autopct="%1.1f%%")
plt.title("Languages User Charts", loc="left")
plt.legend(
           loc='center left',
           bbox_to_anchor=(-0.7, 0.5),
           title='Languages'
           )
plt.tight_layout()
plt.show()
```
